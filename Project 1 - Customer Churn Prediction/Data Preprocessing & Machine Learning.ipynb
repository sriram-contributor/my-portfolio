# Baseline ML Model (Logistic Regression & Decision Tree)

```python
# Split Data into Train & Test Sets
```

```python
from sklearn.model_selection import train_test_split
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

# Hyperparameter Tuning (GridSearchCV & RandomizedSearchCV)

```python
# Tune Random Forest Using GridSearchCV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Define parameter grid
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
# Perform Grd Search
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid_rf, cv=3, scoring='accuracy', n_jobs=-1)
```

```python
grid_search_rf.fit(X_train, y_train)
```

```python
# Best parameters & accuracy
print("Best Perameters:", grid_search_rf.best_params_)
print("Best Accuracy:", grid_search_rf.best_score_)
```

```python
# Train best model
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf_best = best_rf_model.predict(X_test)
```

```python
# Evaluate
print("Tuned Random Forest Performance:\n", classification_report(y_test, y_pred_rf_best))
```

```python
# Tune XGBoost Using RandomizedSearchCV
```

```python
from sklearn.model_selection import RandomizedSearchCV
import xgboost as xgb
```

```python
#Define parameter grid
param_grid_xgb = {
    'n_estimators': [50, 100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}
```

```python
# Perform Random Search
random_search_xgb = RandomizedSearchCV(xgb.XGBClassifier(random_state=42), param_grid_xgb, cv=3, scoring='accuracy', n_iter=10, n_jobs=-1, random_state=42)
```

```python
random_search_xgb.fit(X_train, y_train)
```

```python
# Best parameters & accuracy
print("Best Parameters:", random_search_xgb.best_params_)
print("Best Accuracy:", random_search_xgb.best_score_)
```

```python
# Train best model
best_xgb_model = random_search_xgb.best_estimator_
y_pred_xgb_best = best_xgb_model.predict(X_test)
```

```python
# Evaluate
print("Tuned XGBoost Performance:\n", classification_report(y_test, y_pred_xgb_best))
```
